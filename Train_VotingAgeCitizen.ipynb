```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import scipy
from scipy import stats
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
def WAPE(actual,predicted):
    sum_actual=sum(actual)
    diff=(abs(actual-predicted))
    final_diff=(sum(diff))/sum_actual
    return final_diff*100
```

```python
t1 = pd.read_csv("TrainDataset.csv")
v1 = pd.read_csv("ValidateDataset.csv")
```

```python
vot_actual_t = t1.VotingAgeCitizen
vot_actual_v = v1.VotingAgeCitizen
```

```python
x_train = t1.drop(['Income','VotingAgeCitizen','Employed','PublicWork','Production'],axis='columns')
x_validate = v1.drop(['Income','VotingAgeCitizen','Employed','PublicWork','Production'],axis='columns')
```

```python
model_VotingAgeCitizen= tf.keras.models.Sequential()
model_VotingAgeCitizen.add(tf.keras.layers.Dense(64, tf.keras.activations.relu, input_shape=[len(x_train.keys())]))
model_VotingAgeCitizen.add(tf.keras.layers.Dense(64, tf.keras.activations.relu))
model_VotingAgeCitizen.add(tf.keras.layers.Dense(64, tf.keras.activations.relu))
model_VotingAgeCitizen.add(tf.keras.layers.Dense(1, tf.keras.activations.linear))
```

```python
model_VotingAgeCitizen.compile(optimizer='adam',loss='mean_squared_error',metrics=['mae'])

epoch=50

for i in range(1,8):
    
    print(epoch*i)
    
    model_VotingAgeCitizen.fit(x_train,vot_actual_t, batch_size=1024,epochs=epoch)

    train_ans=np.asarray(model_VotingAgeCitizen.predict(x_train)).reshape(-1,1)
    validate_ans=np.asarray(model_VotingAgeCitizen.predict(x_validate)).reshape(-1,1)

    fig,ax = plt.subplots()
    ax.scatter(vot_actual_t, train_ans)
    ax.plot([vot_actual_t.min(), vot_actual_t.max()], [vot_actual_t.min(), vot_actual_t.max()], 'k-', lw=4)
    ax.set_xlabel('Real Train')
    ax.set_ylabel('Predicted Train')
    plt.show()

    fig,ax = plt.subplots()
    ax.scatter(vot_actual_v, validate_ans)
    ax.plot([vot_actual_v.min(), vot_actual_v.max()], [vot_actual_v.min(), vot_actual_v.max()], 'k-', lw=4)
    ax.set_xlabel('Real Validate')
    ax.set_ylabel('Predicted Validate')
    plt.show()

    print("Train: ",WAPE(np.asarray(vot_actual_t).reshape(-1,1),train_ans))
    print("Validate: ",WAPE(np.asarray(vot_actual_v).reshape(-1,1),validate_ans))
```

```python
test_data = pd.read_csv("TestDataset.csv")
```

```python
final_test = test_data.VotingAgeCitizen
d=test_data.drop(['Income','VotingAgeCitizen','Employed','PublicWork','Production'],axis='columns')
ans2=np.asarray(model_VotingAgeCitizen.predict(d)).reshape(-1,1)
print("Validate: ",WAPE(np.asarray(final_test).reshape(-1,1),ans2))
```
